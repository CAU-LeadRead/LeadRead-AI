```python
import pandas as pd
import numpy as np
import os
import sys
import itertools
```

```python
notes = []
base_dir = "../notes_copy"
df = pd.DataFrame({"brand":[], "name":[], "note":[], "category":[]})
```

```python
def get_notes(dir):
    global df
    text = open(os.path.join(base_dir, dir))
    # lines = file.readlines()
    brand = dir.split(".")[0].replace(' ','').lower()
    count = 0
    while True:
        name = text.readline().lower().strip().replace(" ",'')
        # print(name, len(name))
        # count += 1
        if len(name) == 0:
            break
        top = text.readline().split(":")[-1].strip().lower().split(",")
        mid = text.readline().split(":")[-1].strip().lower().split(",")
        bot = text.readline().split(":")[-1].strip().lower().split(",")
        
        for t in top:
            df_temp = pd.DataFrame({"brand":[brand], "name":[name], "note":[t.strip()], "category":["top"]})
            df = df.append(df_temp,ignore_index=True)
        for m in mid:
            df_temp = pd.DataFrame({"brand":[brand], "name":[name], "note":[m.strip()], "category":["mid"]})
            df = df.append(df_temp,ignore_index=True)
        for b in bot:
            df_temp = pd.DataFrame({"brand":[brand], "name":[name], "note":[b.strip()], "category":["bot"]})
            df = df.append(df_temp,ignore_index=True)

```

```python
get_notes("Acqua di Parma.txt")
get_notes("Byredo.txt")
get_notes("Maison Margiela.txt")
get_notes("Santa Maria Novella.txt")
```

```python
def get_notes2(dir):
    global df
    text = open(os.path.join(base_dir, dir), encoding="utf-8")
    # lines = file.readlines()
    brand = dir.split(".")[0].replace(' ','').lower()
    count = 0
    name = ''
    category = ''
    lines = text.readlines()
    for line in lines:
        if "-" in line:
            name = line.split("-")[-1].lower().strip().replace(" ",'')
        elif "Top Notes" in line:
            category = "top"
        elif "Middle Notes" in line:
            category = "mid"
        elif "Base Notes" in line:
            category = "bot"
        elif len(line.strip()) == 0:
            continue
        else:
            notes = line.strip().lower()
            print('here')
            df_temp = pd.DataFrame({"brand":[brand], "name":[name], "note":[notes], "category":[category]})
            df = df.append(df_temp,ignore_index=True)
        
```
